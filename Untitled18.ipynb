```python
import numpy as np
from scipy.special import betaln, psi

def kl_beta(a1, b1, a0=1, b0=1):
    """
    KL divergence from Beta(a1, b1) to Beta(a0, b0).
    Computes D_KL(Beta(a1,b1) || Beta(a0,b0))
    """
    term1 = betaln(a0, b0) - betaln(a1, b1)
    term2 = (a1 - a0) * psi(a1)
    term3 = (b1 - b0) * psi(b1)
    term4 = (a0 + b0 - a1 - b1) * psi(a1 + b1)
    return term1 + term2 + term3 + term4

# KL divergence for both possible y outcomes
kl_y1 = kl_beta(2, 1)  # Posterior if y = 1
kl_y0 = kl_beta(1, 2)  # Posterior if y = 0

# Since prior is Beta(1,1), the marginal p(y=1) = E[Œ∏] = 0.5
p_y1 = 0.5
p_y0 = 0.5

# Expected KL divergence (in nats)
expected_info = p_y1 * kl_y1 + p_y0 * kl_y0
print(f"Expected information (nats): {expected_info:.4f}")

# Optionally convert to bits
expected_info_bits = expected_info / np.log(2)
print(f"Expected information (bits): {expected_info_bits:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import betaln, psi

def kl_beta(a1, b1, a0=1, b0=1):
    term1 = betaln(a0, b0) - betaln(a1, b1)
    term2 = (a1 - a0) * psi(a1)
    term3 = (b1 - b0) * psi(b1)
    term4 = (a0 + b0 - a1 - b1) * psi(a1 + b1)
    return term1 + term2 + term3 + term4

# Settings
n = 5
n_sim = 10000
theta_true = 0.7  # true theta value used for simulation
kl_vals = []

np.random.seed(42)  # for reproducibility

for _ in range(n_sim):
    y = np.random.binomial(1, theta_true, size=n)
    a_post = 1 + np.sum(y)
    b_post = 1 + n - np.sum(y)
    kl = kl_beta(a_post, b_post)
    kl_vals.append(kl)

kl_vals = np.array(kl_vals)

# Results
print(f"Mean expected information (nats): {kl_vals.mean():.4f}")
print(f"Mean expected information (bits): {kl_vals.mean() / np.log(2):.4f}")

# Plot
plt.hist(kl_vals, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Information Gain over 10,000 Simulations")
plt.xlabel("Information (nats)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
